```python
# import data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns





df=pd.read_csv('mnist_012.csv')

# prepare x and y

X=df.drop(['0'],axis=1)

Y=df.loc[:,'0']
```

```python
#PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=100)
df_pca=pca.fit_transform(X=X)
type(df_pca)
```

```python
#in unsupervised ML model fit_transform happens in a single go
```

```python
# store as dataframe and print

df_pca=pd.DataFrame(df_pca)
print(df_pca.shape)
print('==============')
print('==============')
print(df_pca)
```

```python
df_pca.round(2).head()
```

```python
# variance associated to each PC is obtained 
v=pca.explained_variance_
```

```python
v
```

```python
# this returns the ratio of variance to total variance of the data for each PC

print(pca.explained_variance_ratio_.round(4))
```

```python
# getting a better picture of cumulative variance explained by the 25pcs
```

```python
variances=pca.explained_variance_ratio_.round(4)
percentage=(variances*100)
var=np.cumsum(np.round(variances, decimals=4)*100)
r=np.arange(1,101)
final=[r,variances,percentage,var]
```

```python
final
```

```python
df=pd.DataFrame(data=final)
df1=df.T
columns=['PC#','Variances','Percentage','CUM var']
df1.columns=columns
df1
```

```python
# plotting the scree plot

plt.figure(figsize=(12,6))
pc_values=np.arange(pca.n_components_)+1
plt.plot(pc_values,pca.explained_variance_ratio_,'o-',linewidth=2,color='blue')
plt.title('Scree plot')
plt.xlabel('Principle compo')
plt.ylabel('Var explained')
plt.show()
```

```python
pca=PCA(n_components=50)
df_pca=pca.fit_transform(X=X)

# store as dataframe and print

df_pca=pd.DataFrame(df_pca)
df_pca.round(2).shape
```

```python
variances=pca.explained_variance_ratio_.round(4)
percentage=(variances*100)
var=np.cumsum(np.round(variances, decimals=4)*100)
r=np.arange(1,51)
final=[r,variances,percentage,var]
```

```python
df=pd.DataFrame(data=final)
df1=df.T
columns=['PC#','Variances','Percentage','CUM var']
df1.columns=columns
df1
```

```python
# plotting the scree plot

plt.figure(figsize=(12,6))
pc_values=np.arange(pca.n_components_)+1
plt.plot(pc_values,pca.explained_variance_ratio_,'o-',linewidth=2,color='blue')
plt.title('Scree plot')
plt.xlabel('Principle compo')
plt.ylabel('Var explained')
plt.show()
```
